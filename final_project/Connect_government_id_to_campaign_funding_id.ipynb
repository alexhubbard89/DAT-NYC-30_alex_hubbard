```python
## We need to update the armed_forces_bills to include a unique id to track funding

# Import data to cross reference congress_id (bioguide_id) with fuding_id (opensecrets_id)
all_ids = pd.read_csv('/Users/Alexanderhubbard/Desktop/final_porject_data/all_ids/legislators_all_trancated_rep_only.csv')
all_ids = all_ids.drop('Unnamed: 0', 1)
all_ids.head(2)
```

```python
# create unique list of government_ids
# this iwll be used to search a government id in 'all_ids' and find the opensecrets_id
armed_forces_unique_ids = np.unique(armed_forces_bills['id'])
armed_forces_unique_ids[0]
```

```python
# Add columns to the armed forces bills in order to have an object
# to make the column headers

armed_forces_bills['first_name'] = 'NA'
armed_forces_bills['last_name'] = 'NA'
armed_forces_bills['opensecrets_id'] = 'NA'
col_names = armed_forces_bills.columns
```

```python
# remove the new columns from the armed forces bills
# only need thme to create the col_names object

armed_forces_bills = armed_forces_bills.drop(['first_name', 'last_name', 'opensecrets_id'], 1)
```

```python
## Put congressperson's opensecrets_id and first and last name into armed_forces_bills dataframe
armed_forces_bills_2 = pd.DataFrame(columns=col_names)

# create loop to cycle through all unique congresspeople's bioguide_ids
for i in range(armed_forces_unique_ids.shape[0]):
    start = dt.datetime.now()
    # because some people are missing data we will use the try/exception
    try: 
        # match up bioguide_ids and add appropriate information to armed_forces_bills
        print "{} has info! {} out of {}".format(str(armed_forces_unique_ids[i]), i, armed_forces_unique_ids.shape[0])
        x_get_id_info = all_ids[all_ids['bioguide_id'] == armed_forces_unique_ids[i]]
        x_get_id_info = x_get_id_info.reset_index(drop=True)
        x = armed_forces_bills[armed_forces_bills['id'] == armed_forces_unique_ids[i]]
        x['first_name'] = x_get_id_info['first_name'][0]
        x['last_name'] = x_get_id_info['last_name'][0]
        x['opensecrets_id'] = x_get_id_info['opensecrets_id'][0]
        armed_forces_bills_2 = armed_forces_bills_2.append(x)
    except IndexError:
        print "{} doesn't have info".format(str(armed_forces_unique_ids[0]))
        x_get_id_info = all_ids[all_ids['bioguide_id'] == armed_forces_unique_ids[i]]
        x_get_id_info = x_get_id_info.reset_index(drop=True)
        x = armed_forces_bills[armed_forces_bills['id'] == armed_forces_unique_ids[i]]
        x['first_name'] =  'NA'
        x['last_name'] = 'NA'
        x['opensecrets_id'] = 'NA'
        armed_forces_bills_2 = armed_forces_bills_2.append(x)
    print '{} seconds for {} loop out of {}'.format((dt.datetime.now() - start).seconds,
                                                    i , armed_forces_unique_ids.shape[0])
armed_forces_bills_2.head(2)
```

```python
armed_forces_bills_2 = armed_forces_bills_2.reset_index(drop=True)
```
