Cleaning Data
==

```python
# create structure for master dataframe
bill_colmns = armed_forces_bills.columns
indstry_columns = np.unique(industry_of_interest_contributions['contributor_industry'])
bill_colmns_df = pd.DataFrame(columns=bill_colmns)
industries_to_append = pd.DataFrame(columns=indstry_columns)
master_voting_df = pd.concat([bill_colmns_df, industries_to_append], axis=1)
```

```python
# since everything has already been converted to dats I 
# do not need to in the conditional argument.
# I can also make an object will all unique dates to ease computation
armed_forces_bills_dates = np.unique(armed_forces_bills['date'])

first_date = pd.concat([armed_forces_bills[armed_forces_bills['date'] ==
                                     armed_forces_bills_dates[0]],
                  master_voting_df], axis=0).sort_values(by=['opensecrets_id']).reset_index(drop=True)
id_for_loop = np.unique(first_date['opensecrets_id'])

y = pd.DataFrame(columns=first_date.columns)
for j in range(id_for_loop.shape[0]):
#    print '{} out of {}'.format(j, id_for_loop.shape[0])
    x = first_date[first_date['opensecrets_id'] == id_for_loop[j]]
    for str in range(indstry_columns.shape[0]):
        x[indstry_columns[str]] = industry_of_interest_contributions[(industry_of_interest_contributions['recipient_ext_id'] == 
                                             id_for_loop[j]) &
                                   (industry_of_interest_contributions['contributor_industry'] ==
                                   indstry_columns[str]) &
                                   (industry_of_interest_contributions['date'] <
                                    armed_forces_bills_dates[0])]['amount'].sum()

    y = y.append(x)


print "done!"
y.head()
```

```python
for i in range(armed_forces_bills_dates.shape[0] - 1, armed_forces_bills_dates.shape[0]):
    start = dt.datetime.now()
    print '{} out of {}: {}'.format(i, armed_forces_bills_dates.shape[0], armed_forces_bills_dates[i])
    df_date_loop = pd.concat([armed_forces_bills[armed_forces_bills['date'] ==
                                     armed_forces_bills_dates[i]],
                  master_voting_df], axis=0).sort_values(by=['opensecrets_id']).reset_index(drop=True)
    id_for_loop = np.unique(df_date_loop['opensecrets_id'])
    print '{} id to search'.format(id_for_loop.shape[0])
    for j in range(id_for_loop.shape[0]):
        x = df_date_loop[df_date_loop['opensecrets_id'] == id_for_loop[j]]
        for str in range(indstry_columns.shape[0]):
            person_to_sum = y[y['opensecrets_id'] == id_for_loop[j]].sort('date', ascending=False).reset_index(drop=True)
            if (person_to_sum.shape[0] > 0):
                x[indstry_columns[str]] = (industry_of_interest_contributions[(industry_of_interest_contributions['recipient_ext_id'] == 
                                                                               id_for_loop[j]) &
                                                                              (industry_of_interest_contributions['contributor_industry'] ==
                                                                               indstry_columns[str]) &
                                                                              ((industry_of_interest_contributions['date']) >= 
                                                                               armed_forces_bills_dates[i])]['amount'].sum() + person_to_sum[indstry_columns[str]][0])
            else:
                x[indstry_columns[str]] = industry_of_interest_contributions[(industry_of_interest_contributions['recipient_ext_id'] == 
                                                                               id_for_loop[j]) &
                                                                              (industry_of_interest_contributions['contributor_industry'] ==
                                                                               indstry_columns[str]) &
                                                                              ((industry_of_interest_contributions['date']) >= 
                                                                               armed_forces_bills_dates[i])]['amount'].sum()


        y = y.append(x)
    print '{} seconds: completed {} ids'.format((dt.datetime.now() - start).seconds, id_for_loop.shape[0])
y.head(2)    
```
