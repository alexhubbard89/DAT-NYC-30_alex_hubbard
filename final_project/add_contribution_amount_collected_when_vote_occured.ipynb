Get controbution by date of vote
==

The problem with the original data used was that it looked at contribution data holistically. The block of code below will fix that proble. I am cleaning the data to show how much money a congressperson has accumulated all time up to the point of when a vote took place.

For example, if someone got a 500 dollar contirubtion from the "Misc Defense" industry on 5/1/1990, then the next time they went to vote the data set will show that. I.e. they vote 5/10/1990, the data set will show the bill number, how they voted, and the 500 dollar they have accumulated from the "Misc Defense" industry. Then, if the same person gets an additional contribution for 10,000 dollar form the "Misc Defense" industry on 5/11/1990, then the money will be added to the previos 500 dollar collect for a cumulative total of 10,500 dollar. Therefore, the next time they vote the data set will show the bill number, how the person voted, and 10,500 dollar from the "Misc Defense" industry.

```python
# create structure for master dataframe
bill_colmns = armed_forces_bills.columns
indstry_columns = np.unique(industry_of_interest_contributions_subset['contributor_industry'])
bill_colmns_df = pd.DataFrame(columns=bill_colmns)
industries_to_append = pd.DataFrame(columns=indstry_columns)
master_voting_df = pd.concat([bill_colmns_df, industries_to_append], axis=1)
```

```python
# these are the industries that we are tracking contributions from
indstry_columns
```

First
==

```python
# since everything has already been converted to dats I 
# do not need to in the conditional argument.
# I can also make an object will all unique dates to ease computation
start = dt.datetime.now()
armed_forces_bills_dates = np.unique(armed_forces_bills['date'])

# subset bill data to just the date range we care about
# sort the ids because it makes the loop easier
first_date_bills = pd.concat([armed_forces_bills[armed_forces_bills['date'] ==
                                     armed_forces_bills_dates[0]],
                  master_voting_df], axis=0).sort_values(by=['opensecrets_id']).reset_index(drop=True)
# subset contributions by date range
first_date_contributions = industry_of_interest_contributions_subset[industry_of_interest_contributions_subset['date'] <
                                                                        armed_forces_bills_dates[0]]
# get list of people who participating voting on the date in question
id_for_loop = np.unique(first_date_bills['opensecrets_id'])

# create dataframe to append results/findings
y = pd.DataFrame(columns=first_date_bills.columns)
for j in range(id_for_loop.shape[0]):
    # find person of interst from bill dataframe subset by date range
    x = first_date_bills[first_date_bills['opensecrets_id'] == id_for_loop[j]]
    # find person of interest from contributions dataframe subset by date range
    indy_person_date_range = first_date_contributions[(first_date_contributions['recipient_ext_id'] == 
                                                       id_for_loop[0])]
    # make unique list of industries that contributed to that person for that time
    unique_industries_by_id = np.unique(indy_person_date_range['contributor_industry'])
    for str in range(unique_industries_by_id.shape[0]):
        # use unique industry contributions to find which column to put amount contributed
        x[unique_industries_by_id[str]] = indy_person_date_range[indy_person_date_range['contributor_industry'] == 
                                                                             unique_industries_by_id[str]]['amount'].sum()

    y = y.append(x)


print "done!"
print '{} seconds'.format((dt.datetime.now() - start).seconds)
y.head()
```

Now get data for everyone!
==

```python
for i in range(armed_forces_bills_dates.shape[0]):
    start = dt.datetime.now()
    print '{} out of {}: {}'.format(i, armed_forces_bills_dates.shape[0], armed_forces_bills_dates[i])
    # subset bill data to just the date range we care about
    # sort the ids because it makes the loop easier
    df_date_loop = pd.concat([armed_forces_bills[armed_forces_bills['date'] ==
                                     armed_forces_bills_dates[i]],
                  master_voting_df], axis=0).sort_values(by=['opensecrets_id']).reset_index(drop=True)
    
    # subset contributions by date range
    contributions_df_loop = industry_of_interest_contributions_subset[(industry_of_interest_contributions_subset['date'] <
                                                                               armed_forces_bills_dates[i]) &
                                                                              ((industry_of_interest_contributions_subset['date']) >= 
                                                                               armed_forces_bills_dates[i-1])]
    # get list of people who participating voting on the date in question
    id_for_loop = np.unique(df_date_loop['opensecrets_id'])
    # start loop for each id from that date range
    print '{} id to search'.format(id_for_loop.shape[0])
    for j in range(id_for_loop.shape[0]):
        # find person of interst from bill dataframe subset by date range
        x = df_date_loop[df_date_loop['opensecrets_id'] == id_for_loop[j]]
        # find person of interest from contributions dataframe subset by date range
        indy_person_date_range = contributions_df_loop[(contributions_df_loop['recipient_ext_id'] == 
                                                        id_for_loop[j])]
        # make unique list of industries that contributed to that person for that time
        unique_industries_by_id = np.unique(indy_person_date_range['contributor_industry'])
        # if the person has gather contributions before then we sum contributions they have previously recieved
        # with the contributions they received in this date range
        # a df needs to be created to see if the person has gotten contributions before
        person_to_sum = y[y['opensecrets_id'] == id_for_loop[j]].sort('date', ascending=False).head(1).reset_index(drop=True)
        # start loop of only industries that contributed that person in that date range
        x = x.reset_index(drop=True)
        if person_to_sum.shape[0] > 0:
            x.ix[0, :13] = person_to_sum.ix[0,:13]
        if (person_to_sum.shape[0] > 0):
            # use unique industry contributions to find which column to put amount contributed
            for str in range(unique_industries_by_id.shape[0]):
                if person_to_sum[unique_industries_by_id[str]][0] > 0:
                    num_to_add = person_to_sum[unique_industries_by_id[str]][0]
                    x[unique_industries_by_id[str]] = (indy_person_date_range[indy_person_date_range['contributor_industry'] == 
                                                                              unique_industries_by_id[str]]['amount'].sum() + num_to_add)
                else:
                    num_to_add = 0
                    x[unique_industries_by_id[str]] = (indy_person_date_range[indy_person_date_range['contributor_industry'] == 
                                                                              unique_industries_by_id[str]]['amount'].sum() + num_to_add)    
        else:
            for str in range(unique_industries_by_id.shape[0]):
                x[unique_industries_by_id[str]] = (indy_person_date_range[indy_person_date_range['contributor_industry'] == 
                                                                             unique_industries_by_id[str]]['amount'].sum())
                                                   
        y = y.append(x)
        y = y.reset_index(drop=True)
    print '{} seconds: completed {} ids'.format((dt.datetime.now() - start).seconds, id_for_loop.shape[0])
y.head(2) 
```
